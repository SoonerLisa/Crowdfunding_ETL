```python
#df sub-category.csv project 2

import pandas as pd

crowdfunding_data = pd.read_csv(r'C:\Users\soone\Crowdfunding_ETL\crowdfunding_data.csv')

#Display the first few rows of the df
print(crowdfunding_data.head())
```

```python
#Find the unique categories in the 'sub-category' column:
unique_subcategories = crowdfunding_data['sub-category'].unique()

#Create a new DataFrame to store the mapping of sub-category names to sub-category IDs:
subcategory_mapping = pd.DataFrame({'sub-category': unique_subcategories})

#Add a new column 'sub-category_id' to the sub-category_mapping DataFrame with entries going sequentially from "cat1" to "catn":
subcategory_mapping['sub-category_id'] = 'subcat' + (subcategory_mapping.index + 1).astype(str)

#Merge the original DataFrame crowdfunding_data with the category_mapping DataFrame to add the 'sub-category_id' column based on the 'sub-category' column:
subcategory = crowdfunding_data.merge(subcategory_mapping, on='sub-category', how='left')

subcategory
```
